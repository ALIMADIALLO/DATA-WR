```python
import numpy as np
import pandas as pd
df = pd.read_excel('alima.xlsx')
df.head(12)
```

```python
#mettons l'en tête à partir de la ligne 7
df = pd.read_excel('alima.xlsx', header = 7)
#suppression des lignes et des colonnes vides(axis = 0 ==> pour ligne et axis = 1 ==> pour colonne)
df = df.dropna(how = "all", axis = 1)
df = df.dropna(how = "all", axis = 0).reset_index(drop = True)
df.head(12)
#reset_index pour la redéfinition des index étant donnée les lignes et les colonnes supprimées
```

```python
#pour chaque colonne présente dans le dataframe, ajouter les 5 premières cellules de la colonne si elle ne sont pas vide dans une sous liste qui sera intégrée à la grande liste "columns" pour en faire une liste de liste
columns = []
for c in df.columns:
    columns.append([str(v) for v in df[c][:5] if pd.notnull(v)])
columns
```

```python
#créer une liste columns qui contiendra les en têtes de notre dataframe comme nous le souhaitons

columns = [
    "HDI_rank",
    "Country",
    "Human_poverty_index(HPI-1)-rank",
    "references_1",
    "Human_poverty_index(HPI-1)-value(percent)",
    "Probability_at_birth_of_not_surviving_to_age_40(percent_of_cohort)2000-05",
    "references_2",
    "Adult_illiteracy_rate(percent_aged_15_and_older)1995-2005",
    "references_3",
    "Population_not_using_an_inmproved_water_source(percent)2004",
    "references_4",
    "Children_under_weigth_for_age(percent_under_age_5)1996-2005",
    "references_5",
    "Population_below_income_poverty_line(percent)-one_dollarUS_a_day_1990-2005",
    "references_6",
    "Population_below_income_poverty_line(percent)-two_dollarUS_a_day_1990-2005",
    "references_7",
    "Population_below_income_poverty_line(percent)-National_poverty_line_1990-2004",
    "references_8",
    "HPI-1_rank_minus_income_poverty_rank"
]
df.columns = columns
#suppression de lignes remplacées et réinitialiser les index
df = df.drop(np.arange(5), axis = 0).reset_index(drop = True)
#np.arrange va créer le tableau suivant : [0,1,2,3,4,5] permettant ainsi de supprimer toutes les lignes(axis = 0) possedants les index du tableau
df.head()
```

```python
df.tail(40)
```

```python
#dans cette partie nous allons supprimer toutes les informations qui sont dans les lignes en dessous de notre jeu de donnée c'est à dire après la dernière ligne de données qui nous intéressent
df = df.drop(np.arange(129,len(df)),axis = 0).reset_index(drop = True)
#np.arange va générer un tableau qui commence à la ligne 129 de notre tableau et fini à la fin du dataframe: np.arange(debut(ligne 129), fin(fin de notre fichier))
```

```python
df
```

```python
# transformer la colonne "HDI_rank" en entier(integer)
# Categories
hdi_categories = ["HIGH HUMAN DEVELOPMENT", "MEDIUM HUMAN DEVELOPMENT", "LOW HUMAN DEVELOPMENT"]
# Se débarrasser des rangées avec les catégories HDI
# https://stackoverflow.com/a/27360130 & https://stackoverflow.com/a/12098586
df = df.drop(df[df["HDI_rank"].isin(hdi_categories)].index, axis=0).reset_index(drop=True)
# Référence pour np.where https://docs.scipy.org/doc/numpy/reference/generated/numpy.where.html
# Pensez que cela ressemble à une déclaration du type "if" : where(test est vrai, vrai-quelque chose, faux-quelque chose)
# ceci peut aussi être imbriqué : where(test is true, true-thing, where(new-test is true, true-thing, false-thing))
# transformer la colonne "HDI_rank" en entier(integer)
df["HDI_rank"] = pd.to_numeric(df["HDI_rank"])
df.loc[:, "HDI_category"] = np.where(df["HDI_rank"] > 155,
                                     "LOW HUMAN DEVELOPMENT",
                                     np.where(df["HDI_rank"] > 70,
                                              "MEDIUM HUMAN DEVELOPMENT",
                                              "HIGH HUMAN DEVELOPMENT"
                                             )
                                    )
df.sort_values(by="HDI_rank", inplace=True)
df.head()
```

```python
#supprimer eventuellement les colonnes de references
for i in range(8):
    df = df.drop(columns = f'references_{i+1}', axis = 1)
```
